## 1ï¸âƒ£ If, Elif, Else Statements ðŸ”€

The `if` statement executes code **conditionally** based on boolean expressions.

### Syntax:
```python
if condition1:
    # code block 1
elif condition2:
    # code block 2
else:
    # default code block
```

> ðŸ’¡ **Indentation matters!** Python uses 4 spaces to define code blocks (no curly braces).


````python
# ============================================================
# Basic if statement
# ============================================================
age = 20
if age >= 18:
    print(f"âœ… Age {age}: You are an adult!")
````

````python
# ============================================================
# if-else statement
# ============================================================
temperature = 35
if temperature > 30:
    print(f"ðŸ”¥ {temperature}Â°C â€” It's very hot!")
else:
    print(f"â„ï¸ {temperature}Â°C â€” Nice weather!")
````

````python
# ============================================================
# if-elif-else statement (multiple conditions)
# ============================================================
print("\n--- Grade Calculator ---")
score = 85

if score >= 90:
    grade = "A"
    emoji = "ðŸŒŸ"
elif score >= 80:
    grade = "B"
    emoji = "ðŸ‘"
elif score >= 70:
    grade = "C"
    emoji = "âœ…"
elif score >= 60:
    grade = "D"
    emoji = "âš ï¸"
else:
    grade = "F"
    emoji = "âŒ"

print(f"Score: {score} â†’ Grade: {grade} {emoji}")
````

````python
# ============================================================
# Nested if statements
# ============================================================
print("\n--- Nested If ---")
num = 15
if num > 0:
    print(f"{num} is positive")
    if num % 2 == 0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")
else:
    print(f"{num} is zero or negative")
````

## 2ï¸âƒ£ If-Else One Line (Ternary) & Match Case ðŸŽ¯

### Ternary Operator (One-line if-else):
```python
value = true_result if condition else false_result
```

### Match Case (Python 3.10+):
Similar to `switch` in other languages â€” matches a value against multiple patterns.
```python
match variable:
    case pattern1:
        action1
    case pattern2:
        action2
    case _:
        default_action
```


````python
# ============================================================
# Ternary Operator â€” One-line if-else
# ============================================================
age = 20

# Traditional way
if age >= 18:
    status = "Adult"
else:
    status = "Minor"
````

````python
# Ternary way (same result, one line!)
status = "Adult" if age >= 18 else "Minor"
print(f"Age {age} â†’ {status}")
````

````python
# More examples
x = 10
result = "Even" if x % 2 == 0 else "Odd"
print(f"{x} is {result}")

grade = "Pass" if 75 >= 60 else "Fail"
print(f"Grade: {grade}")
````

````python
# Nested ternary (use sparingly â€” can be hard to read!)
score = 85
grade = "A" if score >= 90 else "B" if score >= 80 else "C" if score >= 70 else "F"
print(f"Score {score} â†’ Grade {grade}")
````

````python
# ============================================================
# Match Case (Python 3.10+) â€” Structural Pattern Matching
# ============================================================
print("\n--- Match Case ---")

def get_day_type(day):
    """Determine if a day is a weekday or weekend using match-case."""
    match day.lower():
        case "saturday" | "sunday":
            return "ðŸ–ï¸ Weekend"
        case "monday" | "tuesday" | "wednesday" | "thursday" | "friday":
            return "ðŸ’¼ Weekday"
        case _:
            return "â“ Unknown day"

for day in ["Monday", "Saturday", "Holiday"]:
    print(f"{day:>10} â†’ {get_day_type(day)}")
````

````python
# Match with patterns
print("\n--- Match with Patterns ---")
def describe_number(num):
    match num:
        case 0:
            return "zero"
        case n if n > 0:
            return "positive"
        case n if n < 0:
            return "negative"

for n in [0, 42, -7]:
    print(f"{n:>3} â†’ {describe_number(n)}")
````
